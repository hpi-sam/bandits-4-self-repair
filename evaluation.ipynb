```python
import os
import pandas as pd
import itertools
import seaborn
import math
from matplotlib import pyplot
from scipy import std, mean
from scipy.stats import mode, kurtosis, skew
from numpy import median
import random
from sklearn.metrics import mean_squared_error
from math import sqrt
import numpy as np
import statistics
import pickle
```

```python
ucb_trials = pickle.load(open('ucb_trials.p', 'rb'))
epsilon_greedy_trials = pickle.load(open('epsilon_greedy_trials.p', 'rb'))
```

```python

regrets = []
for index in range(len(ucb_trials)):
    params = 'params'
    epsilon = 'epsilon'
    decay = 'decay'
    regrets.append({'regret': ucb_trials[index]['regret'], 'algorithm': 'ucb', 'params': f"epsilon: {ucb_trials[index][params][epsilon]} decay: {ucb_trials[index][params][decay]}"})
    regrets.append({'regret': epsilon_greedy_trials[index]['regret'], 'algorithm': 'eps', 'params': f"epsilon: {epsilon_greedy_trials[index][params][epsilon]} decay: {epsilon_greedy_trials[index][params][decay]}"})   

#regrets = [f'ucb_regret{index}': ucb_trials[index]['regret'], f'eps_regret{index}': epsilon_greedy_trials[index]['regret'] for index in range(len(ucb_trials))]
pyplot.figure()
pyplot.subplots(figsize=(20, 10))
seaborn.barplot(data=pd.DataFrame(regrets), x='params', y='regret', hue='algorithm')
pyplot.show()
```

```python

percentages = []
for index in range(len(ucb_trials)):
    params = 'params'
    epsilon = 'epsilon'
    decay = 'decay'
    percentages.append({'optimal_arm_percentage': ucb_trials[index]['percentage_optimal_arm_used'], 'algorithm': 'ucb', 'params': f"epsilon: {ucb_trials[index][params][epsilon]} decay: {ucb_trials[index][params][decay]}"})
    percentages.append({'optimal_arm_percentage': epsilon_greedy_trials[index]['percentage_optimal_arm_used'], 'algorithm': 'eps', 'params': f"epsilon: {epsilon_greedy_trials[index][params][epsilon]} decay: {epsilon_greedy_trials[index][params][decay]}"})   

#regrets = [f'ucb_regret{index}': ucb_trials[index]['regret'], f'eps_regret{index}': epsilon_greedy_trials[index]['regret'] for index in range(len(ucb_trials))]
pyplot.figure()
pyplot.subplots(figsize=(20, 10))
seaborn.barplot(data=pd.DataFrame(percentages), x='params', y='optimal_arm_percentage', hue='algorithm')
pyplot.show()
```

```python

iteration_counts = []
for index in range(len(ucb_trials)):
    params = 'params'
    epsilon = 'epsilon'
    decay = 'decay'
    iteration_counts.append({'iterations': ucb_trials[index]['iterations'], 'algorithm': 'ucb', 'params': f"epsilon: {ucb_trials[index][params][epsilon]} decay: {ucb_trials[index][params][decay]}"})
    iteration_counts.append({'iterations': epsilon_greedy_trials[index]['iterations'], 'algorithm': 'eps', 'params': f"epsilon: {epsilon_greedy_trials[index][params][epsilon]} decay: {epsilon_greedy_trials[index][params][decay]}"})   

#regrets = [f'ucb_regret{index}': ucb_trials[index]['regret'], f'eps_regret{index}': epsilon_greedy_trials[index]['regret'] for index in range(len(ucb_trials))]
pyplot.figure()
pyplot.subplots(figsize=(20, 10))
seaborn.barplot(data=pd.DataFrame(iteration_counts), x='params', y='iterations', hue='algorithm')
pyplot.show()
```
